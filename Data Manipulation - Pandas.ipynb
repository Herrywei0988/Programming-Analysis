```python
#Run this cell first before moving on
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
%matplotlib inline
```

## COVID19 Cases DataFrame

```python
"""Problem 1"""

#Use the variable/url to load the data
# url for covid cases:  
covid_cases = "https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
# It is okay to have a horizontal scroll bar for this cell.
df_cases = pd.read_csv(covid_cases)
df_cases
```

#### Remove/Drop DataFrame Columns

```python
"""Problem 2"""
# Remove the following columns for df_cases using the drop method: 
# "UID", "iso2", "iso3", "code3", "FIPS", "Country_Region", "Lat", "Long_"
# Set the inplace=True.

df_cases.drop(columns=["UID", "iso2", "iso3", "code3", "FIPS", "Country_Region",
                       "Lat", "Long_"], inplace = True)   #remove --- and enter code
df_cases.head()
```

#### Rename DataFrame Columns

```python
"""Problem 3"""
# Rename the following columns for df_cases dataframe using the rename method:
# 'Admin2' to 'county', 'Province_State' to 'state', 'Combined_Key' to 'county_state'
# Set inplace=True

df_cases.rename(columns={'Admin2' : 'county', 'Province_State' : 'state',
                         'Combined_Key' : 'county_state'}, inplace= True)

#Run cell to make sure the columns are renamed by using the head method
print(df_cases.head())
```

#### Melt/Pivot DataFrame Columns

```python
df_cases
```

```python
"""Problem 4"""
#Melt/reshape df_cases. Assign the following when using pd.melt()
# Columns to assign to the id_vars parameter are: 'county', 'state', 'county_state'
# Column name to assign to var_name parameter: "date" 
# Column name to assign value_name parameter: "cases" 
# Complete the melt function code below.

df_cases_melted = pd.melt(df_cases, id_vars=['county', 'state', 'county_state'], var_name="date", value_name="cases")
df_cases_melted.tail(3)
```

```python
df_cases_melted
```

```python
df_cases_melted.dtypes
```

#### Convert Date Columns to DateTime Data Type

```python
"""Problem 5"""
# Using df_cases_melted, change data type for date to a datetime object. 

df_cases_melted.date = pd.to_datetime(df_cases_melted.date)
print(df_cases_melted.dtypes)
```

## COVID19 Deaths DataFrame


```python
"""Problem 6"""
# Use the variable/url to load the data
# url for covid deaths:  
covid_deaths = "https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv"
# It is okay to have a horizontal scroll bar for this cell.
df_deaths = pd.read_csv(covid_deaths)
df_deaths

# Remove the following columns for df_deaths using the drop method: 
# "UID", "iso2", "iso3", "code3", "FIPS", "Country_Region", "Lat", "Long_"
# Set the inplace=True.

df_deaths.drop(columns=["UID", "iso2", "iso3", "code3", "FIPS", "Country_Region",
                        "Lat", "Long_", "Population"], inplace = True)
```

```python
"""Problem 7"""
# Rename the following columns for df_deaths dataframe using the rename method:
# 'Admin2' to 'county', 'Province_State' to 'state', 'Combined_Key' to 'county_state'
# Set inplace=True

df_deaths.rename(columns={'Admin2' : 'county', 'Province_State' : 'state',
                         'Combined_Key' : 'county_state'}, inplace= True)
```

```python
"""Problem 8"""
# View the first five rows in df_deaths - make sure the columns have been renamed

df_deaths.head()
```

```python
"""Problem 9"""
# Melt/reshape df_deaths. Assign the following when using pd.melt()  
# Columns to assign to the id_vars parameter are: 'county', 'state', 'county_state'
# Column name to assign to var_name parameter: "date" 
# Column name to assign value_name parameter: "deaths" 


df_deaths_melted = pd.melt(df_deaths, id_vars=['county', 'state', 'county_state'], var_name='date', value_name='deaths')

df_deaths_melted.tail(3)
```

```python
"""Problem 10"""
# Using df_deaths_melted, change data type for date to a datetime object. 

df_deaths_melted.date = date = pd.to_datetime(df_deaths_melted.date)
print(df_deaths_melted.dtypes)
```

### Merge/Join DataFrames

```python
"""Problem 11"""
#Merge the melted dataframes for cases and deaths.
#Complete the merged function code below.

df_merged = pd.merge(df_cases_melted, df_deaths_melted)
df_merged.tail()
```

### Date Calculations 

```python
df_merged
```

```python
df_merged.date
```

```python
"""Problem 12"""
# Calculate the number of days into the outbreak for each record. 
# The name of the new column will be us_outbreak

df_merged["us_outbreak"] = df_merged.date - df_merged.date.min();


# The following code outputs the number of days since the outbreak (Day 0)
print(df_merged[["date", "us_outbreak"]].iloc[-1])
```

```python
# Run cell to view a sample of the dataframe with the new data column (us_outbreak)
df_merged
```

### Aggregate the DataFrame

```python
"""
Problem 13
After completing merge, run the code below and review the output.
Understand what the as_index parameter does.
"""
df = df_merged.groupby(["us_outbreak", "date"], 
                       as_index=False)[["cases", "deaths"]].sum()
df
```

### Derived Data

```python
"""Problem 14"""
#Using df from above, create a derived data column using the diff method on cases.
#The name of the new column is: new_cases

df["new_cases"] = df.cases.diff()
df.tail()
```

```python
df["new_deaths"] = df.deaths.diff()
df
```

### Data Visualization 

```python
""" Problem 15 """
#Use plt.bar() function to create a bar plot,
# have the dates on the x-axis and new_cases on the y-axis.
#Provide x & y axis labels and an appropriate chart title that
#  explains visual. 
#Rotate the x-axis ticks to 45 using the xticks function.
# Below are the functions used - complete the code:

plt.figure(figsize=(10, 8)) # include figsize keyword argument
plt.bar(df.date, df.new_cases)
plt.xticks(rotation=45) # include rotation keyword argument
plt.xlabel("Date")
plt.ylabel("Cases (In million)")
plt.title("New Cases of COVID-19")
plt.tight_layout()
plt.show()
```
