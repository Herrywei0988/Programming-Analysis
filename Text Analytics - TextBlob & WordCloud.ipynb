#### Sentiment Analysis with TextBlob

In sentiment analysis, the TextBlob polarity score is a float that ranges from -1 to 1. <br> 
- A score of -1 indicates a negative sentiment, while a score of 1 indicates a positive sentiment. <br> 
- A score of 0 indicates a neutral sentiment. The polarity score is calculated by taking the average of the sentiment scores of all the words in the text.

The TextBlob polarity score is a simple and effective way to measure the sentiment of text. However, it is important to note that it is not always perfect. For example, the score may not accurately reflect the sentiment of text that is ironic or sarcastic. Additionally, the score may be skewed by words that are not well-represented in the TextBlob lexicon.

Despite its limitations, the TextBlob polarity score is a useful tool for understanding the overall sentiment of text. It can be used to compare the sentiment of different pieces of text, or to track the sentiment of a piece of text over time.

Here are some examples of how the TextBlob polarity score can be used:

A company can use the TextBlob polarity score to analyze customer feedback. This can help the company identify areas where it can improve its products or services.
A politician can use the TextBlob polarity score to analyze the sentiment of social media posts about them. This can help the politician understand how their policies are being perceived by the public.
A journalist can use the TextBlob polarity score to analyze the tone of news articles. This can help the journalist identify articles that are biased or misleading.
The TextBlob polarity score is a versatile tool that can be used to gain insights from a variety of types of text

### NaiveBayesAnalyzer() 
The NaiveBayesAnalyzer is a sentiment analyzer that is included in the TextBlob library. It is a classification approach that has been trained on a dataset of movie reviews. The NaiveBayesAnalyzer returns a namedtuple of the form Sentiment(classification, p_pos, p_neg), where:

classification is the sentiment of the text (either "positive" or "negative")
p_pos is the probability that the text is positive
p_neg is the probability that the text is negative

```python
# 3. View the sentiment using an alternative train model, NaiveBayesAnalyzer, in blob2 variable.

from textblob.sentiments import NaiveBayesAnalyzer

blob2 = TextBlob(text, analyzer = NaiveBayesAnalyzer())
blob2.sentiment
```
