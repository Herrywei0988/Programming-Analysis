### Import libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use("ggplot")
```

### Load data 

```python
# Load data tips data from seaborn
df = sns.load_dataset("tips")
print(df.head())
```

```python
# inspect data using info() method
df.info()
```

### Histogram 

```python
#1 
# Create a histogram showing the distribution of total bill for all customers.

plt.hist(x = df.total_bill)
plt.title("Total Bill (All Customer)")
plt.xlabel("Set Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#2 
# Create a histogram showing the distribution of total bill for females.

female = df[df.sex =="Female"]
plt.hist(female.total_bill)
plt.title("Total Bill (Female)")
plt.xlabel("Bill Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#3 
# Create a histogram showing the distribution of total bill for males.

male = df[df.sex =="Male"]
plt.hist(male.total_bill)
plt.title("Total Bill (Male)")
plt.xlabel("Bill Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#4 
# Create one figure showing two histograms that overlap displaying the 
#   total bill distribution for males and females, respectively.  
# Include the keyword argument alpha=0.5 for transparency.

plt.hist(male.total_bill, alpha = 0.5, label = "Male")
plt.hist(female.total_bill, alpha = 0.5, label = "Female")
plt.legend()
plt.show()
```

### Box Plot

```python
#1 
# Create a box plot showing the distribution of tips for all customers. 

plt.boxplot(df.tip)
plt.xticks([])
plt.title("Customer Tip")
plt.ylabel("Tip Amount")
plt.show()
```

```python
#2
# Create two box plots using one figure with one axes showing a comparison of tips for females and males.

plt.boxplot([female.tip, male.tip])
plt.xticks(ticks = [1, 2], labels= ["Females", "Male"])
plt.title("Customer Tips")
plt.ylabel("Tip Amount")
plt.show
```

### One Figure / Multiple Subplots

```python
# Create a figure containing 4 subplots in a 2 rows by 2 columns layout
# first subplot: a histogram of total bill for all customers (histogram #1 above)
# second subplot: two overlapping hisograms of total bill for males and females (histogram #4 above)
# third subplot: a box plot on tips for all customers (box plot #1 above)
# fourth subplot: two box plots comparing female and male tips  (box plot #2 above)

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(15, 10))

ax[0, 0].hist(df.total_bill)
ax[0, 0].set_xlabel("Bill Amount")
ax[0, 0].set_ylabel("Frequency")
ax[0, 0].set_title("Total Bill (All Customer)")
#plt.hist(df.total_bill)
#plt.xlabel("Bill Amount")
#plt.ylabel("Frequency")
#plt.title("Total Bill (All Customers)")
#plt.show


ax[1, 0].boxplot(df.tip)

plt.show()
```

### Strip Plot

```python
# Create a strip plot using seaborn that shows male and female customers and their tips. 

sns.stripplot(x= df.sex, y= df.tip)

plt.show()
```

### Scatter Plot

```python
# Create a scatter plot showing the relationship between total bill and tips for all customers

sns.scatterplot(x= "total_bill", y = "tip", data = df)
plt.show()
```
