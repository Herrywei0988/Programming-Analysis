**Loading Data:**
- How do you read a CSV file into a Pandas DataFrame?
- What are some other common file formats you can read into a DataFrame?
- How can you view the first few rows of a DataFrame to inspect its contents?


```python
url = "https://raw.githubusercontent.com/babdelfa/dataq/main/tips.csv"

import pandas as pd

tips = pd.read_csv(url)
tips.head(3)
```

```python
tips.size
```

**Data Inspection:**
- How do you check the number of rows and columns in a DataFrame?
- What is the purpose of the `info()` method in Pandas?
- How can you see summary statistics for numerical columns?

```python
print(tips.shape)
print(tips.info())
print(tips.describe())
```

**Data Selection:**
- How do you select a specific column from a DataFrame?
- How can you select multiple columns at once?
- What is the difference between using square brackets and dot notation to select columns?

```python
tip = tips["tip"]
print(tip)

tips[["tip", "total_bill"]]
```

**Filtering Data:**
- How do you filter rows in a DataFrame based on a condition?
- What is a boolean mask, and how can it be used for filtering?

```python
sat = tips[tips.day == "Sat"]
sat
```

**Data Aggregation and Grouping:**
- What is the `groupby()` function used for in Pandas?
- How can you calculate summary statistics for groups of data (e.g., mean, sum, count)?
- How do you reset the index of a grouped DataFrame?

```python
average_tips = tips.groupby("sex")["tip"].mean()
average_tips.plot(kind="bar")
```
