```python
import random

def generate_random_sequence(length, possible_values):
    return [random.choice(possible_values) for _ in range(length)]

def evaluate_guess(target, guess):
    result = ""
    used_indices = set()
    
    for i in range(len(guess)):
        if guess[i] == target[i]:
            result += "●"
            used_indices.add(i)
    
    for i in range(len(guess)):
        if guess[i] != target[i] and guess[i] in target and i not in used_indices:
            result += "○"
            used_indices.add(i)
    
    return result

def is_valid_guess(guess, possible_values):
    return len(guess) == len(possible_values) and all(char in possible_values for char in guess)

def play_mastermind():
    possible_values = ["1", "2", "3", "4", "5", "6"]
    target_sequence = generate_random_sequence(4, possible_values)
    max_turns = 12
    turns = 0

    print("Welcome to Mastermind!")
    
    while turns < max_turns:
        guess = input(f"Turn {turns + 1}/{max_turns}: Enter your 4-value guess (e.g., 1234): ")
        
        if not is_valid_guess(guess, possible_values):
            print("Invalid guess. Please enter a 4-value sequence using values 1-6.")
            continue
        
        result = evaluate_guess(target_sequence, guess)
        print(f"Result: {result}")
        
        if result == "●●●●":
            print("Congratulations! You've won!")
            break
        
        turns += 1
    
    if turns == max_turns:
        print("Sorry, you've run out of turns. The correct sequence was:", "".join(target_sequence))

if __name__ == "__main__":
    play_mastermind()
```

```python
import random

def generate_random_sequence():
    return [random.randint(1, 6) for _ in range(4)]

def evaluate_guess(secret_sequence, guess):
    result = ""
    used_indices = set()
    
    # Check for exact matches (●)
    for i in range(4):
        if guess[i] == secret_sequence[i]:
            result += "●"
            used_indices.add(i)
    
    # Check for partial matches (○)
    for i in range(4):
        if guess[i] != secret_sequence[i] and guess[i] in secret_sequence and i not in used_indices:
            result += "○"
            used_indices.add(i)
    
    return result

def main():
    secret_sequence = generate_random_sequence()
    turns_left = 12
    
    print("Welcome to Mastermind!")
    print("Try to guess the 4-value sequence with numbers from 1 to 6.")
    
    while turns_left > 0:
        print(f"Turns left: {turns_left}")
        guess = input("Enter your 4-value guess: ")
        
        # Validate input
        if len(guess) != 4 or not all(char.isdigit() and 1 <= int(char) <= 6 for char in guess):
            print("Invalid input. Please enter exactly 4 values in the range of 1-6.")
            continue
        
        guess = [int(char) for char in guess]
        
        result = evaluate_guess(secret_sequence, guess)
        print(f"Result: {result}")
        
        if result == "●●●●":
            print("Congratulations! You guessed the correct sequence!")
            break
        
        turns_left -= 1
    
    if turns_left == 0:
        print("You ran out of turns. The correct sequence was:", secret_sequence)

if __name__ == "__main__":
    main()
```
