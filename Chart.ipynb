```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/babdelfa/class/main/EmployeeSalaries.csv")

fig, axes = plt.subplots(nrows =2,ncols=2, figsize=(9, 7))
plt.suptitle('Analysis of Employee Salaries')

axes[0, 0].hist(df.Salary, bins=10, color = "gray")
axes[0, 0].set_title("All Employee Salaries")
axes[0, 0].set_xlabel("Salary", fontsize = 8)
axes[0, 0].set_ylabel("Frequency" , fontsize = 8)

##

yes = df[df.MBA == "Yes"]
no = df[df.MBA == "No"]

axes[0, 1].set_title("Salary Comparison 1")
axes[0, 1].set_xlabel("Salary", fontsize=8)
axes[0, 1].set_ylabel("Frequency", fontsize=8)
axes[0, 1].hist(yes.Salary, color = 'green', alpha = 0.2, label = "Yes")
axes[0, 1].hist(no.Salary,color = 'red', alpha = 0.2, label = "No")
axes[0, 1].legend(title="MBA")

##

axes[1, 0].set_title("Salary Comparison 2")
axes[1, 0].set_xlabel("Age", fontsize=8)
axes[1, 0].set_ylabel("Salary", fontsize=8)
sns.scatterplot(x="Age", y="Salary", data=df, hue="MBA", palette={"No": "red", "Yes":"green"}, ax=axes[1,0])
plt.tight_layout()

##

axes[1, 1].set_title("Salary Comparison 3")
axes[1, 1].set_xlabel("MBA", fontsize = 8)
axes[1, 1].set_ylabel("Salary" , fontsize = 8)
sns.stripplot(x= df.MBA, y= df.Salary, palette={"No": "red", "Yes":"green"})

plt.tight_layout(pad = 2.0)
_ =plt.show
```
