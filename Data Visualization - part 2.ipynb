```python
# Task 0: Import libaries and load data frame 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style
plt.style.use("ggplot")

# Load the CSV data file
df = pd.read_csv(r"https://raw.githubusercontent.com/babdelfa/class/main/EmployeeSalaries.csv")  
# Display a sample of the dataframe
print(df.head())
```

```python
# Task : Develop a seaborn boxplot between Age and Salary

# sns.catplot()
sns.boxplot(x= df.MBA, y = df.Salary)
```

```python
# Task 2: Develop a seaborn jointplot between Age and Salary

sns.jointplot(x = "Age", y = "Salary", data = df)
```

```python
# Task 3: Develop a seaborn stripplot comparing salaries of MBA and non-MBA employees

sns.stripplot(x = df.MBA, y = df.Salary)
plt.show()
```

```python
# Task 4: Develop a seaborn scatterplot showing the relationship between Age and Salary

sns.scatterplot(x= "Age", y = "Salary", data = df)
plt.show()
```

```python
# Task 5 Create a 1x2 subplot figure (complete the code)
fig, axes = plt.subplots(2, 1, figsize=(6, 6))

# Task 3 subplot :
sns.stripplot() #use Task 3 code 


# Task 4 subplot:
sns.scatterplot(x="Age", y="Salary", data=df, hue="MBA", ax=axes[1]) #used Task 4 code 
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Salary")
axes[1].set_title("Age vs Salary (Colored by MBA Status)")


# Adjust layout
plt.tight_layout()
plt.show()
```
